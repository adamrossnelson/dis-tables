```python
import pandas as pd
import numpy as np
from scipy.stats import f
from markdownify import markdownify
```

```python
f.ppf(1-.05, 2, 12)
```

```python
df = pd.DataFrame(np.zeros(1500).reshape(100,15), 
                  columns=['dfBtwn{}'.format(i) 
                           for i in range(1,16)],
                  index=['dfWthn{}'.format(i) 
                         for i in range(1,101)])
```

```python
df.head()
```

```python
for b in range(0,15):
    for w in range(0,100):
        df.iloc[w,b] = f.ppf(1-.05, b+1, w+1)
```

```python
df.head(n=15)
```

```python
df.to_html('Generated_F_Table.html')
```

```python
file = open('Generated_F_Table.html', 'r').read()
html = markdownify(file)

mkdn = open('Generated_F_Table.md', 'w')
mkdn.writelines('# Generated_F_Table.md')
mkdn.writelines(html)
mkdn.close()
```
