```python
import pandas as pd
import numpy as np
from scipy.stats import t
```

```python
t.ppf(1-.05, 10)
```

```python
df = pd.DataFrame(np.zeros(500).reshape(100,5), 
                  columns=['Alpha {}'.format(i) 
                           for i in [.1, .05, .025, .01, .001]],
                  index=['df{}'.format(i) 
                         for i in range(1,101)])
```

```python
df.head()
```

```python
for a, k in zip([.1, .05, .025, .01, .001], range(0,5)):
    for w in range(0,100):
        #print('{} {}'.format(w, k))
        df.iloc[w,k] = t.ppf(1-a/2, w+1)
```
